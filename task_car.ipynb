```python
import random
trials = 1000
count_sum_7 = 0
for _ in range(trials):
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    
    if die1 + die2 == 7:
        count_sum_7 += 1
experimental_probability = count_sum_7 / trials
count_sum_7
experimental_probability
```

```python
import random
trials = 10000
count = 0
for _ in range(trials):
    coin = random.choice(["H", "T"])
    die = random.randint(1, 6)
    
    if coin == "H" and die == 6:
        count += 1
count / trials
```
